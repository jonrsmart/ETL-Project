```python
Hotels_file = '../cleaned/good_cheap_airbb.csv'
Hotels_df = pd.read_csv(Hotels_file)
Hotels_df.head()
```

```python
#Transform Hotels DataFrame
```

```python
# Create a filtered dataframe from specific columns
Hotels_cols = ["listing_url", "name", "description"]
Hotels_transformed= Hotels_df[Hotels_cols].copy()

# Rename the column headers
Hotels_transformed = Hotels_transformed.rename(columns={"listing_url": "Hotel listing_url",
                                                          "name": "Hotel_name",
                                                          "description": "Room_description"})

Hotels_df.drop(columns=['picture_url',
                        'neighbourhood_group_cleansed',
                        'has_availability',
                        'review_scores_accuracy',
                        'review_scores_value',
                        'instant_bookable',
                        'reviews_per_month',], inplace=True)
                    






Hotels_transformed.head()
```
